### 3b. Tuning-GridSearchCV

```python
def cv_score(clf, x, y, score_func = accuracy_score):
    result = 0
    nfold = 5
    x.reset_index(drop = True, inplace = True)
    y.index = x.index
    for train, test in KFold(nfold, shuffle = True,
                             random_state = 42).split(x):
        clf.fit(x.loc[train, :], y[train])
        result += score_func(clf.predict(x.loc[test, :]), y[test])
    return result / nfold
```

```python
clf = LogisticRegression(solver = 'liblinear', max_iter = 500)
score = cv_score(clf, X_train, y_train)
print(f'{score:.2f}')
```

```python
scores = []
Cs = [0.001, 0.1, 1, 10, 100]
for c in Cs:
    clf = LogisticRegression(C=c)
    score = cv_score(clf, X_train, y_train)
    scores.append(score)
    print("The average score for C-Value {} is {}".format(c, score))
```

```python
clf = LogisticRegression(solver = 'liblinear', max_iter = 10 )
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(accuracy_score(y_pred, y_test))
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator = LogisticRegression(), param_grid = {'C': [0.001, 0.1, 1, 10, 100]}, cv=6, scoring="accuracy")
grid.fit(X_train, y_train)
y_pred = grid.predict(X_test)
#print(accuracy_score(y_pred, y_test))
grid.best_estimator_, grid.best_params_, grid.best_score_
```
