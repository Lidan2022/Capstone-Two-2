```python
List=['LocationAbbr', 'Data_Value_Unit', 'SubMeasure', 'TopicType', 'MeasureDesc', 'DataSource', 'Response', 'TopicTypeId', 'TopicId', 'MeasureId', 'StratificationID1','StratificationID2', 'StratificationID3', 'StratificationID4', 'SubMeasureID']
df1.drop(columns=List, inplace = True)
df1.head()
```

```python
pd.concat([df1['Data_Value_Footnote_Symbol'],
           df1['Data_Value_Footnote'], 
           df1['Gender'], df1['Age'], df1['Education'], df1['TopicDesc']]).unique()
```

```python
df1.drop(columns=['Data_Value_Type', 'Data_Value_Footnote_Symbol', 'Data_Value_Footnote', 'Age'], inplace=True)
df1.head()
```

```python
df1.rename(columns={'YEAR':'year', 'LocationDesc':'state', 'TopicDesc':'tabacco_type','Data_Value':'percentage_tabacco_user', 'Gender':'gender', 'Race':'race', 'Education':'education'}, inplace=True)
df1.head()
```

```python
List=['LocationAbbr', 'Data_Value_Unit', 'TopicType', 'MeasureDesc', 'DataSource', 'Response', 'TopicTypeId', 'TopicId', 'MeasureId', 'StratificationID1','StratificationID2', 'StratificationID3', 'StratificationID4', 'SubMeasureID']
df2.drop(columns=List, inplace = True)
df2.head()
```

```python
pd.concat([df2['Data_Value_Footnote_Symbol'],
           df2['Data_Value_Footnote'], 
           df2['Gender'], df2['Age'], df2['Education'], df2['TopicDesc']]).unique()
```

```python
df2.drop(columns=['Data_Value_Type', 'Data_Value_Footnote_Symbol', 'Data_Value_Footnote', 'Age'], inplace=True)
df2.head()
```

```python
df2.rename(columns={'YEAR':'year', 'LocationDesc':'state', 'TopicDesc':'tabacco_type','Data_Value':'percentage_tabacco_user', 'Gender':'gender', 'Race':'race', 'Education':'education'}, inplace=True)
df2.head()
```

```python
df1.isnull().sum()
```

```python
cols = ['percentage_tabacco_user','Data_Value_Std_Err', 'Low_Confidence_Limit', 'High_Confidence_Limit', 'Sample_Size']
df1[cols] = df1[cols].fillna(df1[cols].mean())
df1.isnull().sum()
```

```python
cols = ['percentage_tabacco_user','Data_Value_Std_Err', 'Low_Confidence_Limit', 'High_Confidence_Limit', 'Sample_Size']
df2[cols] = df2[cols].fillna(df2[cols].mean())
df2.isnull().sum()
```

```python
df2['GeoLocation'].fillna('ffill', inplace=True)
df2.isna().sum()
```

```python
df2.duplicated()
```

```python
df = pd.concat([df1,df2])
df
```
